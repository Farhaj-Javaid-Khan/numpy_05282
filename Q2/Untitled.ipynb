```python
from PIL import Image
image = Image.open('ffff.jpg')
print(image.format)
print(image.size)
print(image.mode)
```

```python
from matplotlib import image
from matplotlib import pyplot
image = image.imread('ffff.jpg')
print(image.dtype)
print(image.shape)
pyplot.imshow(image)
pyplot.show()
```

```python
from PIL import Image
from numpy import asarray

image = Image.open('ffff.jpg')

data = asarray(image)
print(type(data))

print(data.shape)


image2 = Image.fromarray(data)
print(type(image2))


print(image2.mode)
print(image2.size)
```
